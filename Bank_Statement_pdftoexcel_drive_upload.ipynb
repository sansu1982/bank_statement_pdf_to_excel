```python
# Install necessary libraries
!pip install pdfplumber
!pip install PyPDF2

import pdfplumber
import PyPDF2
import pandas as pd
from getpass import getpass  # For secure password input
from google.colab import drive

# Mount Google Drive (if PDF stored there)
drive.mount('/content/drive')

# File paths
pdf_path = "/content/drive/MyDrive/Colab Notebooks/Bank Statement PDF - Excel/Axis-01-04-2018to31-03-2019.pdf"
output_excel = "bank_statement.xlsx"

# Ask for password at runtime (not stored in file)
pdf_password = getpass("Enter PDF password: ")

# Step 1: Check and decrypt if needed
with open(pdf_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)
    if reader.is_encrypted:
        try:
            reader.decrypt(pdf_password)
        except:
            raise Exception("Incorrect password or unable to decrypt PDF.")

# Step 2: Extract tables from PDF using pdfplumber
rows = []
with pdfplumber.open(pdf_path, password=pdf_password) as pdf:
    for page in pdf.pages:
        tables = page.extract_tables()
        for table in tables:
            for row in table:
                rows.append(row)

# Step 3: Save to Excel
df = pd.DataFrame(rows)
df.to_excel(output_excel, index=False)

print(f"✅ Saved extracted data to {output_excel}")
```
