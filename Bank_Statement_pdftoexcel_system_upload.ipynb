Conversion after file is uploaded from the system

```python
# Install dependencies
!pip install pdfplumber
!pip install PyPDF2

import pdfplumber
import PyPDF2
import pandas as pd
from google.colab import files
from getpass import getpass

# Step 1: Upload PDF file from local system
uploaded = files.upload()

# Step 2: Get the uploaded file path
pdf_path = list(uploaded.keys())[0]
output_excel = "bank_statement.xlsx"

# Step 3: Securely get password (hidden while typing)
pdf_password = getpass("Enter PDF password: ")

# Step 4: Check and decrypt if needed
with open(pdf_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)
    if reader.is_encrypted:
        try:
            reader.decrypt(pdf_password)
        except:
            raise Exception("Incorrect password or unable to decrypt PDF.")

# Step 5: Extract data using pdfplumber
rows = []
with pdfplumber.open(pdf_path, password=pdf_password) as pdf:
    for page in pdf.pages:
        tables = page.extract_tables()
        for table in tables:
            for row in table:
                rows.append(row)

# Step 6: Save to Excel
df = pd.DataFrame(rows)
df.to_excel(output_excel, index=False)

print(f"✅ Saved extracted data to {output_excel}")

# Step 7: Download Excel file
files.download(output_excel)
```
